### Purpose:

* <span>The purpose of this final project is to assess how income level, population size, greenhouse gas emissions and electricity demand influence the share of fossil fuel versus renewable electricity generation across countries.</span>
* <span>The **key dependent variables** are:
  1. Fossil fuel electricity share
  2. Renewable energy electricity share </span>
* <span>The **key independent variables** are:
  1. Low-income
  2. Lower-middle income
  3. Upper-middle income
  4. High-income
  5. Population
  6. Electricity demand
  7. Greenhouse gas emissions</span>
* <span>This dataset contains a total of 23,196 cases.
* <span>Each case represents either a country, region or income-level grouping for a specific year.
* <span>Because the dataset spans multiple decades, the number of cases is much larger than the number of countries.</span>

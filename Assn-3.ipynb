```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

# enter additional commands here to reproduce the figure below

x = np.arange(len(a)) # self-note: create x values that match the number of given data points; to make sure each point is plotted in order 

plt.plot(x, a, marker = 'o', linestyle = '-', color = 'black') # line and dots # a is the vertical values (loaded data)
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

# NOTE!
# - Entering commands and running the cell may cause the figure to disapper
# - If the figure disappears, use Assn-3.html as a reference
```

### Figure 2

Use `data-fig2.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig2.npy')
x = np.arange(len(a))

plt.plot(x, a[: ,0], label = 'Line 1 (Blue)') # all the rows in column 1 # give each line a name using the label = argument
plt.plot(x, a[: ,1], label = 'Line 2 (Orange)') # all the rows in column 2
plt.plot(x, a[:,2], label = 'Line 3 (Green)') # column 3

plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig3.npy')
# x = np.arange(len(a)) both x and y are given so no need for this 
# print(a.shape)

x = a[:, 0]
y = a[:, 1]

plt.scatter(x, y, s=200, color = 'blue', alpha = 0.4) #size = 200 and alpha is transparent points
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

### Figure 4

Use `data-fig4.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig4.npy')
# print(a.shape)
x = np.arange(len(a))

plt.bar(x, a, color = 'grey')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig5.npy')
# print(a.shape)

plt.imshow(a, cmap = 'RdBu', origin = 'lower', aspect = 'auto') # main visualisation command for heatmap; origin makes points start at bottom-left corner, aspect makes sure image fits space naturally 
plt.colorbar()
plt.show()
```
